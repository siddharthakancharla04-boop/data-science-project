```python

import matplotlib.pyplot as plt

# Years from 2021 to 2026
years = [2021, 2022, 2023, 2024, 2025, 2026]

# Example sales data
sales = [1200, 1800, 2500, 3200, 4000, 4800]

plt.plot(years, sales, marker='o')

plt.title("EV Sales Growth (2021-2026)")
plt.xlabel("Year")
plt.ylabel("Sales Volume")

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [2021, 2022, 2023, 2024, 2025, 2026]
sales = [1200, 1800, 2500, 3200, 4000, 4800]

plt.bar(years, sales, color='blue')

plt.title("EV Sales Comparison (2021-2026)")
plt.xlabel("Year")
plt.ylabel("Sales Volume")

plt.show()
```

```python
import matplotlib.pyplot as plt

years = ['2021', '2022', '2023', '2024', '2025', '2026']
sales = [1200, 1800, 2500, 3200, 4000, 4800]

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']

plt.pie(sales, labels=years, autopct='%1.1f%%', colors=colors)

centre_circle = plt.Circle((0, 0), 0.60, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("EV Sales Donut Chart (20212026)")
plt.show()
```

```python
plt.fill_between(years, sales, alpha=0.4)

plt.title("EV Sales Area Chart (2021-2026)")
plt.xlabel("Year")
plt.ylabel("Sales Volume")

plt.show()
```

```python
plt.hist(sales, bins=5)

plt.title("Sales Distribution (2021-2026)")
plt.xlabel("Sales Range")
plt.ylabel("Frequency")

plt.show()
```
